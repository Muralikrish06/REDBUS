```python
driver=webdriver.Chrome()
wait=WebDriverWait(driver,30)
A=driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")
all_data=[]
def scrape_page():
    routescontainer = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "route_link")))
    for route in routescontainer:
        try:
            routename=route.find_element(By.CSS_SELECTOR,"a[class='route']").text
            routelink=route.find_element(By.CLASS_NAME,'route').get_attribute('href')
            all_data.append({'routename':routename,'routelink':routelink})
        except Exception as e:
            print(f"An error occurred: {e}")
            continue
for page_number in range(1,4):
    scrape_page()
    if page_number < 2:
        try:
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class, "DC_117_pageTabs") and text()="{page_number + 1}"]')
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(1)
            print(f"Clicking on page {page_number + 1}")
            next_page_button.click()
            wait.until(EC.text_to_be_present_in_element((By.XPATH, '//div[contains(@class, "DC_117_pageTabs DC_117_pageActive")]'), str(page_number + 1)))
            print(f"Successfully navigated to page {page_number + 1}")
            time.sleep(3)
        except Exception as e:
            print(f"An error occurred while navigating to page {page_number + 1}: {e}")
            break        

           

df_AP=pd.DataFrame(all_data)
df_AP.to_csv("all_data.csv",index=False)
display(df_AP)
                   
Bus_data=[]

for i,j in df_AP.iterrows():
    routename=j["routename"]
    routelink=j["routelink"]
    driver.get(routelink)
    time.sleep(5)
    try :
        b=driver.find_elements(By.CLASS_NAME,'button')
        
        time.sleep(5)
        for i in range(len(b)-1,-1,-1):
            b[i].click()
    except Exception as e:
        print(f"An error occurred: {e}") 
    
    
    while True:

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            last_height = driver.execute_script("return document.body.scrollHeight")
        # Scroll down to the bottom in order to load all the buses

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for page to load
            time.sleep(2)

        # Calculate new scroll height and compare with last scroll height

            new_height = driver.execute_script("return document.body.scrollHeight")

        # If it is the same height then it is at the end of the page
            if new_height == last_height:
                break

            last_height = new_height
                 
    try:
        bus_container=wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"bus-item")))
        time.sleep(5)
        
        for bus in bus_container:
            
            try:

                    busname=bus.find_element(By.CLASS_NAME,"travels ").text
                    bustype=bus.find_element(By.CLASS_NAME,"bus-type ").text
                    price=bus.find_element(By.CLASS_NAME,'fare').text
                    seat_available=bus.find_element(By.XPATH,".//div[contains(@class, 'seat-left')]").text
                    depature_time=bus.find_element(By.CLASS_NAME,"dp-time ").text
                    travel_time=bus.find_element(By.CLASS_NAME,"dur").text
                    end_time=bus.find_element(By.CLASS_NAME,"bp-time ").text
                    star_rating=bus.find_element(By.CLASS_NAME,'rating-sec').text
                    Bus_data.append({'busname':busname,'bustype':bustype,'depature_time':depature_time,'travel_time':travel_time,'end_time':end_time,'star_rating':star_rating,'price':price,'seat_available':seat_available,'routename':routename,'routelink':routelink})
            except Exception as e:
                    print(f"Error scraping bus data: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")
BUS_ANDRA=pd.DataFrame(Bus_data)
BUS_ANDRA.to_csv("Bus_data.csv",index=False)
driver.quit()
display(BUS_ANDRA)
```
