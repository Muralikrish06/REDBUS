```python
A["price"]=A["price"].str.replace("INR","")
A["price"]=A["price"].astype(float)
A
```

```python
A["star_rating"]=A["star_rating"].str.replace("New",'')
A["star_rating"]=pd.to_numeric(A["star_rating"],errors='coerce')


```

```python
A["seat_available"]=A["seat_available"].str.replace("Seats available",'')
A["seat_available"]=A["seat_available"].str.replace("seat available",'')
A["seat_available"]=pd.to_numeric(A["seat_available"],errors='coerce')
A["seatavilable"]=A["seat_available"].fillna(0).astype(int)
```
